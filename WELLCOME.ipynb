```python
def journal_title_mapper(j):
    if j == 'AMERICAL JOURNAL OF PSYCHIATRY':
        return 'AMERICAN JOURNAL OF PSYCHIATRY'
    else:
        return j
```

```python
df['Journal title'] = df['Journal title'].apply(lambda x: journal_title_mapper(x))
```

```python
df['Journal title'] = df['Journal title'].str.replace('FOR', 'OF').str.strip()
```

```python
def journal_title_mapper1(i):
    if i == 'JOURNAL OF VISUALIZED EXPERIEMENTS':
        return 'JOURNAL OF VISUALIZED EXPERIMENTS'
    else:
        return i
```
